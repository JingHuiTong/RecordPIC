```python
import pandas as pd
import numpy as np
import os
import glob
```

```python
path = 'Repeat_10-8'


cluster = (path.rsplit('_')[-1]).rsplit('-')[0]
TopN    = (path.rsplit('_')[-1]).rsplit('-')[1]
```

```python
for dirpath in glob.glob('Repeat_10-8'):
    cluster = (dirpath.rsplit('_')[-1]).rsplit('-')[0]
    TopN    = (dirpath.rsplit('_')[-1]).rsplit('-')[1]
    for csvpath in glob.glob(f'{dirpath}/*/*_Cluster_?.csv'):
        print(csvpath)
        category = (csvpath.rsplit('/')[1])
        print(category)
        df = pd.read_csv(csvpath)

        drank = {'name':[],
                'num':[] }
        for cc in range(int(cluster)):
            df_name = df[f'clusterName_{cc}'].tolist()
            df_num  = df[f'clusterNum_{cc}'].tolist()
            for _name, _num in zip(df_name,df_num):
                if pd.isnull(_name) != True :
                    if not  (_name in drank['name']):
                        drank['name'].append(_name)
                        drank['num'].append(int(_num))

                    else:

                        idx = drank['name'].index(_name)
                        orinum = drank['num'][idx]
                        drank['num'][idx] =int(orinum+_num)

        drank=pd.DataFrame(drank)
        drank=drank.sort_values(by='num',ascending=False)
        drank.to_csv(f'{dirpath}/{category}/{category}_clear_rank.csv',index=False)
    
```
